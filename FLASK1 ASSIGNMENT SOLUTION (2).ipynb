Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a lightweight and flexible web framework for Python. It is designed to be simple and easy to use, yet powerful enough to build a wide range of web applications.

* Advantages of Flask Framework:

1. Simplicity: Flask has a simple and intuitive syntax, making it easy to get started and learn.

2. Lightweight: Flask is a lightweight framework with a small core and modular design.
3. Flexibility: Flask provides great flexibility, allowing developers to choose the components and extensions they need for their specific project requirements.
